### Forward Selection (AIC)

```python
import warnings

# Suppress all Perfect Separation and Convergence Warnings
warnings.filterwarnings("ignore", message=".*Perfect separation.*")
warnings.filterwarnings("ignore", message=".*Maximum Likelihood optimization failed to converge.*")
```

```python
import pandas as pd
import statsmodels.api as sm

df_encoded = pd.get_dummies(df.drop('churn', axis=1), drop_first=True)

y = df['churn']
X_all = df_encoded.copy()

def forward_selection(X, y):
    selected_features = []
    remaining_features = list(X.columns)
    current_score, best_new_score = float('inf'), float('inf')

    while remaining_features:
        aic_with_candidates = []

        for candidate in remaining_features:
            features_to_test = selected_features + [candidate]
            X_model = sm.add_constant(X[features_to_test])
            try:
                model = sm.Logit(y, X_model).fit(disp=0)
                aic = model.aic
                aic_with_candidates.append((aic, candidate))
            except Exception as e:
                continue

        if not aic_with_candidates:
            break

        aic_with_candidates.sort()
        best_new_score, best_candidate = aic_with_candidates[0]

        if best_new_score < current_score:
            remaining_features.remove(best_candidate)
            selected_features.append(best_candidate)
            current_score = best_new_score
            print(f" Added: {best_candidate} | AIC: {round(current_score, 2)}")
        else:
            break

    print("\n Final selected features (Forward):")
    print(selected_features)
    return selected_features

print("=== Forward Selection ===")
selected_forward = forward_selection(X_all, y)
```

### 4. Feature Selection via Forward Selection (AIC)

- **Observation:** The forward selection method based on AIC incrementally added 16 features that contributed to reducing the model’s AIC score. The process began with `current_month_debit` and included both numerical and encoded categorical variables.
- **Insight:** The most impactful features selected include `current_balance`, `average_monthly_balance_prevQ`, and `branch_code`, indicating their strong predictive value for churn. Additionally, categorical encodings such as `occupation_self_employed`, `gender_Female`, and `gender_Male` were also included, highlighting their relevance in the model. The selection process prioritized features that contributed significantly to model fit and avoided overfitting by stopping when AIC no longer improved.

### Backward Elimination (AIC)

```python
import pandas as pd
import statsmodels.api as sm
import numpy as np

# Assuming df and df_encoded are already defined
# Ensure 'churn' is binary and cleaned
y = df['churn']

# Ensure X is all numeric
X_all = df_encoded.copy()
X_all = X_all.apply(pd.to_numeric, errors='coerce')

# Drop rows with any missing values (NaNs introduced by coercion or already present)
X_all = X_all.dropna()
y_cleaned = y.loc[X_all.index]

# Additional checks for types
print("\nChecking dtypes and unique values before backward elimination...")

# Check if there are any non-numeric columns
print("\nX_cleaned dtypes (showing non-numeric):")
print(X_all.dtypes[X_all.dtypes != 'float64'])

# Ensure the target variable is binary
print("\nUnique target values in y:")
print(y_cleaned.unique())

# Check if there are any object columns in X
if any(X_all.dtypes == 'object'):
    print("\nSome columns still have object dtype.")
    print(X_all.dtypes[X_all.dtypes == 'object'])
else:
    print("\nAll features are numeric.")

# Ensure X and y are both numeric
X_all = X_all.astype(float)
y_cleaned = y_cleaned.astype(float)

# Function for backward elimination
def backward_elimination(X, y):
    selected_features = list(X.columns)

    # Initial model fit
    try:
        initial_model = sm.Logit(y, sm.add_constant(X[selected_features])).fit(disp=0)
        current_score = initial_model.aic
    except Exception as e:
        print("Initial model failed:", e)
        return []

    while len(selected_features) > 0:
        aic_with_candidates = []

        for candidate in selected_features:
            features_to_test = [f for f in selected_features if f != candidate]
            try:
                X_model = sm.add_constant(X[features_to_test])
                model = sm.Logit(y, X_model).fit(disp=0)
                aic = model.aic
                aic_with_candidates.append((aic, candidate))
            except Exception as e:
                print(f"Skipped {candidate} due to error: {e}")
                continue

        if not aic_with_candidates:
            break

        # Sort by the lowest AIC
        aic_with_candidates.sort()
        best_new_score, worst_feature = aic_with_candidates[0]

        # Remove the worst feature if AIC improves
        if best_new_score < current_score:
            selected_features.remove(worst_feature)
            current_score = best_new_score
            print(f"Removed: {worst_feature} | AIC: {round(current_score, 2)}")
        else:
            break

    print("\nFinal selected features (Backward):")
    print(selected_features)
    return selected_features

# Run backward elimination
print("\n=== Backward Elimination ===")
selected_backward = backward_elimination(X_all, y_cleaned)
```

### 5. Feature Selection via Backward Elimination (AIC)

During this process, 7 features were eliminated, including both numerical and categorical variables that had minimal impact on model performance.

- **Insight**:  
  The final set of 16 selected features retained key variables such as `current_balance`, `current_month_credit`, and `previous_month_debit`, reflecting their importance in predicting churn.
